# Metodo congrencial multiplicativo

Los generadores congruenciales multiplicativos, también conocidos como generadores de números aleatorios de Lehmer, son un tipo de generador congruencial lineal para generar números pseudoaleatorios⁴. 

No pude encontrar la fecha exacta de su creación, pero son uno de los métodos más antiguos y conocidos para la generación de números pseudoaleatorios¹².

En cuanto a su uso actual, aunque los Generadores Congruenciales Lineales Multiplicativos (GLMC) son capaces de producir números pseudoaleatorios que pueden pasar las pruebas de aleatoriedad, esta posibilidad es extremadamente sensible a la elección de los parámetros¹. Por lo tanto, no deberían ser usados en aplicaciones para las que se requiera aleatoriedad de alta calidad¹. Sin embargo, el GLC sigue siendo lo suficientemente bueno para tareas sencillas como la prueba de primalidad de Miller-Rabin, o los repartos de FreeCell⁴.

En cuanto a las ventajas y desventajas:
- Ventajas: Los GLC son rápidos, y requieren poca memoria (normalmente 32 o 64 bits) para retener su estado. Esto los hace valiosos para simular flujos múltiples¹. Además, se puede reproducir fácilmente una secuencia de números, a partir de la misma semilla⁴.
- Desventajas: En una secuencia dada por un generador congruencial cualquiera, los puntos están ubicados en no más de hiperplanos paralelos⁷. Además, el Generador Congruencial Lineal Multiplicativo (GLMC) no cumple la primera condición, pues no tienen periodo completo¹.

# Generador Mersenne Twistter

El Mersenne Twister es un generador de números pseudoaleatorios de propósito general desarrollado en 1997 por Makoto Matsumoto y Takuji Nishimura¹. Su nombre proviene del hecho de que la longitud de su período se elige para ser un número primo de Mersenne¹. Fue diseñado específicamente para rectificar la mayoría de los defectos encontrados en los generadores de números pseudoaleatorios más antiguos¹.

La versión más comúnmente utilizada del algoritmo Mersenne Twister se basa en el número primo de Mersenne. La implementación estándar de eso, MT19937, utiliza una longitud de palabra de 32 bits¹. Existe otra implementación (con cinco variantes¹) que utiliza una longitud de palabra de 64 bits, MT19937-64; genera una secuencia diferente¹.

El Mersenne Twister se utiliza como generador de números pseudoaleatorios predeterminado en varios lenguajes de programación y software, incluyendo Dyalog APL, IDL, R, Ruby, Free Pascal, PHP, Python, CMU Common Lisp, Embeddable Common Lisp, Steel Bank Common Lisp, Julia, GLib, GNU Multiple Precision Arithmetic Library, GNU Octave, GNU Scientific Library, Microsoft Excel, GAUSS, gretl, Stata, SageMath, Scilab, Maple, MATLAB, entre otros¹.

El Mersenne Twister tiene varias ventajas:
- Licencia permisiva y libre de patentes para todas las variantes excepto CryptMT¹.
- Pasa numerosas pruebas de aleatoriedad estadística, incluyendo las pruebas Diehard y la mayoría, pero no todas, las pruebas TestU01.
- Un período muy largo de $$2^{19937} - 1$$.
- Las implementaciones generalmente crean números aleatorios más rápido que los métodos implementados en hardware¹.

El Mersenne Twister sigue siendo ampliamente utilizado en varios lenguajes de programación y software, incluyendo Dyalog APL, IDL, R, Ruby, Free Pascal, PHP, Python, CMU Common Lisp, Embeddable Common Lisp, Steel Bank Common Lisp, Julia, GLib, GNU Multiple Precision Arithmetic Library, GNU Octave, GNU Scientific Library, Microsoft Excel, GAUSS, gretl, Stata, SageMath, Scilab, Maple, MATLAB, entre otros¹.

Aquí están algunas de las ventajas y desventajas del Mersenne Twister:

**Ventajas**:
- Licencia permisiva y libre de patentes para todas las variantes excepto CryptMT.
- Pasa numerosas pruebas de aleatoriedad estadística, incluyendo las pruebas Diehard y la mayoría, pero no todas, las pruebas TestU01.
- Un período muy largo de $$2^{19937} - 1$$.
- Las implementaciones generalmente crean números aleatorios más rápido que los métodos implementados en hardware.

**Desventajas**:
- Puede llevar mucho tiempo convertir un estado inicial no aleatorio (notablemente la presencia de muchos ceros) en una salida que pase las pruebas de aleatoriedad.
- La semilla no debe ser 0.
- A pesar de su gran tamaño de estado, falla en algunas pruebas estadísticas estándar.
- Tiene problemas serios alrededor de 0, lo que sugiere que se aleatoriza bastante mal.
- En comparación con los RNGs simples como XorShift*, es también inesperadamente complicado.

```python
class metodosNumerosAleatorios:
    def __init__(self, size: int):
        """metodosNumerosAleatorios
        Esta clase tiene implementado los siguientes algoritmos para la creación de numeros pseudo aleatorios, he aqui cuales estan implementados:
        
        - generadorGreen
        - generadorTausworthe
        - congruential_generator
        - generadorCongruencialCuadratico
        - congruenciaLinealMultiplicativo
        - Rand
        - Randu
        - congruenciaLineal
        - cuadrados_medios
        - random_number_self
        
        
        Y otros metodos:
        - comprobar_numeros_dos: comprueba cuantas veces y en que indices se repiten los numeros
        - pruebaDeMedia

        Args:
            size (int): Cantidad de numeros a generar.
        """
        self.size = size
        # self.semilla = semilla
        # self.numeros = list()
        # self.a = a
        # self.c = c
        # self.m = m

    def random_number_self(self, normalizar : bool, guardar = False)-> list[int,]:

        if self.size > 0:
            if not normalizar:
                nums = list()
                for i in range(1, self.size + 1):
                    nums.append((((self.size**i) // i) + (self.size - i)) % self.size)
                    
                if guardar:
                    with open(file='./pruebas numeros/algoritmo_propio_parametros_n_'+str(self.size)+'_normalizado_' + str(guardar)+'.txt', mode='w') as f:
                        for item in nums:
                            f.write(str(item))
                            f.write('\n')
                            
                return nums
            else:
                nums = list()
                for i in range(1, self.size + 1):
                    nums.append(((((self.size**i) // i) + (self.size - i)) % self.size) / self.size)
                    
                if guardar:
                    with open(file='./pruebas numeros/algoritmo_propio_parametros_n_'+str(self.size)+'_normalizado_' + str(guardar)+'.txt', mode='w') as f:
                        for item in nums:
                            f.write(str(item))
                            f.write('\n')
                            
                return nums
        else:
            return []

    def cuadrados_medios(self, x_0 : int, normalizar = False, guardar = False):
        """Funcion que retorna numeros aleatorios del tamaño que se indica en la función, utilizando el algoritmo de cuadrados medios

        Args:
            size (int): Cantidad de numeros aleatorios a generar
            semilla (int, optional): Semilla que se requiere para generar los números aleatorios, tiene que ser de más de 3 digitos. Defaults to 333.

        Returns:
            list[int,]: Lista con la cantidad de numeros pseudoaleatorios que se indican.
        """
        t = len(str(x_0))
        # assert t >= 3, "El tamaño de la semilla es incorrecto"
        if t >= 3:
        
            if not normalizar:
                nums = [x_0]

                for i in range(self.size):
                    num = str(nums[i] ** 2)
                    dif = (len(num) - t) // 2

                    n = num[dif : t + dif]

                    if len(n) % 2 == 1:
                        n = "0" + n
                    n = int(n)

                    nums.append(n)
                    
                if guardar:
                    with open(file='./pruebas numeros/algoritmo_cuadrados_medios_n_'+str(self.size)+'_semilla_' + str(x_0)+ '_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                        for item in nums:
                            f.write(str(item))
                            f.write('\n')
                
                return nums[1:]
            else:
                nums = [x_0]

                for i in range(self.size):
                    num = str(nums[i] ** 2)
                    dif = (len(num) - t) // 2
                    n = num[dif : t + dif]

                    if len(n) % 2 == 1:
                        n = "0" + n
                    n = int(n)

                    nums.append(n)
                    
                nums = np.array(nums[1:])
                nums = (nums - np.min(nums)) / (np.max(nums) - np.min(nums))
                
                if guardar:
                    with open(file='./pruebas numeros/algoritmo_cuadrados_medios_n_'+str(self.size)+'_semilla_' + str(x_0)+ '_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                        for item in nums:
                            f.write(str(item))
                            f.write('\n')
                            
                return nums
        else:
            return []

    def congruenciaLineal(self, xn: int, a: int, c:int, m: int, normalizar=False, guardar = False)-> list[int]:
        """Función para generar numeros aleatorios usando el algoritmo de congruencia lineal. Este método genera una secuancia de números pseudoaletorios a partir de una semilla inicial, utilizando la fórmula:
        xn+1 = (a * xn + c) mod m
        Args:
            xn (int): Valor inicial o semilla
            a (int): Valor multiplicativo de la formula
            c (int): Valor aditivo de la formula
            m (int): Valor por el que se va a obtener el modulo
            normalizar (bool, optional): Cuando esta opcion es verdadera regresa valores entre [0, 1]. Defaults to False.
            guardar (bool, optional): Cuando esta opcion es verdadera guarda los numeros en un archivo de texto. Defaults to False.

        Returns:
            list[int,]: Arreglo con la cantidad de números generados aleatoriamente
        """
        nums = [xn]

        if not normalizar:
            for i in range(self.size):
                xn1 = (a * nums[i] + c) % m
                nums.append(xn1)
        else:
            for i in range(self.size):
                xn1 = (a * nums[i] + c) % m
                nums.append(xn1/ m)

        if guardar:
            with open(file='./pruebas numeros/algoritmo_congruencia_lineal_n_'+str(self.size)+'_semilla_' + str(xn)+ 'valor_a_' + str(a)+'_valor)c_' + str(c) + '_valor_m_'+str(m) +'_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                for item in nums[1:]:
                    f.write(str(item))
                    f.write('\n')
        return nums[1:]

    def Randu(self, x_0 : int, m: int, normalizar=False, guardar = False):
        """Randu Generador de número aleatorios utiilzado en los 60's y 70's se define como:
        Xn+1 = (2*16 + 3) Xn mod (2*31)

        Args:
            x_0 (int): Valor inicial, también llamado semilla
            m (int): Valor por el cual se va a obtener el modulo
            normalizar (bool, optional): _description_. Defaults to False.
            guardar (bool, optional): _description_. Defaults to False.

        Returns:
            _type_: _description_
        """
        assert self.size > 0, "El número ingresado no es valido"
        assert x_0 > 0, "El número ingresado no es valido"

        nums = [x_0]
        # m = 2**31

        if not normalizar:
            for i in range(self.size):
                nums.append(((2**16) + 3) * nums[i] % m)
        else:
            for i in range(self.size):
                nums.append((((2**16) + 3) * nums[i] % m) / m)
                
        if guardar:
            with open(file='./pruebas numeros/algoritmo_Randu_n_'+str(self.size)+'_semilla_' + str(x_0)+ '_valor_m_'+str(m) +'_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                for item in nums[1:]:
                    f.write(str(item))
                    f.write('\n')

        return nums[1:]

    def Rand(self, m : int, x_0 : int, normalizar=False, guardar = False):
        """Rand Por muchos años (antes de 1995) el generador de la función rand en matlab fue el generador congruencial:
        xn+1 = (7**5) * xn mod(2**31 -1)

        Args:
            m (int) : Valor por el cual se va a obtener el modulo
            x_n (int): Valor inicial, también llamado semilla
            normalizar (bool, optional): Cuando este valor es positivo regresa valores en un rango de [0, 1]. Defaults to False.
            guardar (bool, optional): Cuando este valor es positivo guarda los numeros generados en un archivo de texto plano. Defaults to False

        Returns:
            list[int,]: Arreglo con la cantidad de numeros generados
        """
        nums = [x_0]

        if not normalizar:
            for i in range(self.size):
                nums.append((7**5) * nums[i] % m)
        else:
            for i in range(self.size):
                nums.append((((7**5) * nums[i]) % m) / m)
                
        if guardar:
            with open(file='./pruebas numeros/algoritmo_Rand_n_'+str(self.size)+'_semilla_' + str(x_0)+ '_valor_m_'+str(m) +'_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                for item in nums[1:]:
                    f.write(str(item))
                    f.write('\n')
                    
        return nums[1:]

    def comprobar_numeros_dos(self, ubi="./Numeros generados/"):
        # Comprobar si los archivos ya existen
        if not os.path.exists(ubi):
            os.makedirs(ubi)
            self.generar_numeros()

        numeros_unicos = set()
        indices = {}

        # Leer los archivos y realizar el seguimiento de los índices
        for file in os.listdir(ubi):
            if not file.startswith("resultados_dos"):  # or file != 'resultados.txt':
                with open(os.path.join(ubi, file), "r") as f:
                    for index, element in enumerate(f):
                        n = float(element)
                        if n.is_integer():
                            n = int(n)
                        else:
                            n = round(n)  # Redondear números no enteros
                        numeros_unicos.add(n)
                        indices.setdefault(n, []).append(index)

        # Calcular las frecuencias
        frecuencias = {numero: indices[numero] for numero in numeros_unicos}

        # Escribir los resultados en un archivo
        with open(
            os.path.join(ubi, "resultados_dos_" + str(self.parameters) + ".txt"),
            "w",
            encoding="utf-8",
        ) as file:
            for numero, index_list in frecuencias.items():
                file.write(
                    f"El número {numero} aparece un total de {len(index_list)} veces, en los índices: {index_list}\n"
                )

        print("Terminado")
        
    def congruenciaLinealMultiplicativo(self, x_0 : int, a: int, m: int, normalizar = False, guardar = False)-> list[int,]:
        """Variante del algorimo de congruencia lineal sigue la formula:
        x_n+1 = a * x_n % m

        Args:
            x_0 (int): Valor inicial, tambien llamado semilla
            a (int): valor multiplicativo
            m (int): valor por el que se va a obtener el residuo
            normalizar (bool, optional): Cuando este valor sea positivo va a regresar valores en un rango de [0, 1]. Defaults to False.
            guardar (bool, optional): Cuando este valor sea positivo va a guardar los numeros generares en un archivo de texto plano. Defaults to False.

        Returns:
            list[int,]: _description_
        """
        nums = [x_0]
        
        if not normalizar:        
            for i in range(self.size):
                nums.append((a * nums[i]) % m)
        else:
            for i in range(self.size):
                nums.append(((a * nums[i]) % m) / m)
                
        if guardar:
            with open(file='./pruebas numeros/algoritmo_congruenciaLinealMultiplicativo_n_'+str(self.size)+'_semilla_' + str(x_0)+ '_valor_a_'+str(a) +'_valor_m_'+str(m) +'_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                for item in nums[1:]:
                    f.write(str(item))
                    f.write('\n')
            
        
        return nums[1:]
    
    def generadorCongruencialCuadratico(self, a_1 : int, a_2 : int, x_0 : int,  c : int, m: int, guardar = False, normalizar =False)-> list[int,]:
        """Generador Congruencial cuadratico:
        Basado en la formula:
            X_n+1 = (a_1 * X_n ** 2 + a_2 * x_n + c) % m
        Donde el periodo maximo es m

        Args:
            a_1 (int): Valor multiplicativo del valor cuadratico.
            a_2 (int): valor multiplicativo del valor lineal.
            x_0 (int): Valor inicial también llamado semilla.
            c (int): Valor aditivo de la formula
            m (int): Valor por el cual se va a obtener el residuo, también llamado rango.
            guardar (bool): Valor que cuando es positvo guarda los numeros en un archivo de texto plano. Default False
            normalizar (bool): Valor boleano que cuando es positivo retorna valores en un rango [0, 1]

        Returns:
            list[int,]: Arreglo que contiene los numeros generados aleatoriamente
        """
        # assert if 
        nums = [x_0]
        if not normalizar:
            for i in range(self.size):
                nums.append(((a_1 * (nums[i] **2)) + (a_2 * nums[i]) + c) % m)
        else:
            for i in range(self.size):
                nums.append((((a_1 * (nums[i] **2)) + (a_2 * nums[i]) + c) % m) / m)
                
        if guardar:
            with open(file='./pruebas numeros/algoritmo_generadorCongruencialCuadratico_n_'+str(self.size)+'_semilla_' + str(x_0)+ '_valor_a1_'+str(a_1)+ '_valor_a2_'+str(a_2) +'_valor_m_'+str(m) +'_normalizado_'+str(normalizar)+'.txt', mode='w') as f:
                for item in nums[1:]:
                    f.write(str(item))
                    f.write('\n')
        
        return nums[1:]
    
    def congruential_generator(self,a : list, X : list, m : int, n : int) ->list[int,]:
        """Generador congruencial generalizado

        Args:
            a (list): lista de coeficientes
            X (list): lista de semillas iniciales
            m (int): módulo
            n (int): número de números pseudoaleatorios a generar. no importa.

        Returns:
            list[int,]: lista con la cantidad de numeros especificados.
        """
        # a: lista de coeficientes
        # X: lista de semillas iniciales
        # m: módulo
        # n: número de números pseudoaleatorios a generar

        k = len(a)  # longitud de la lista de coeficientes
        assert len(X) == k, "La longitud de la lista de semillas debe ser igual a la longitud de la lista de coeficientes"

        # Generar la secuencia
        for _ in range(self.size):
            X_next = sum(a[i]*X[-(i+1)] for i in range(k)) % m
            X.append(X_next)

        # Devolver los números pseudoaleatorios generados
        return X[k:]
    
    def generadorTausworthe(self, a:list, x: int, n:int) -> list[int,]:
        """Generador Tausworthe
        Basado en la formula:
        X_n = (a_1*X_n-1 + a_2 * X_n-2+ ... + a_k * X_n-k) % 2
        
        Sirve para generar secuencias de bit

        Args:
            a (list): Lista de coeficientes
            x (int): Lista de semillas
            n (int): Cantidad de numeros a generar

        Returns:
            list[int,]: Retorna una lista con la cantidad de números aleatorios
        """
        k = len(a)
        assert len(x) == k, "La longitud de la lista de semillas debe ser igual a la longitud de la lista de coeficientes"
        
        for _ in range(n):
            x_next = sum(a[i]*x[i] for i in range(k)) % 2
            x.append(x_next)
            
        return x[k:]
    
    def generadorGreen(self, x:list, m:int, n : int)->list[int,]:
        
        k = len(x)
        for _ in range(n):
            x_next = (x[-1] + x[-k]) % m
            x.append(x_next)
            
        return x[k:]
    
    def pruebaDeMedia(self, numeros:list, valor_z : float):
        """pruebaDeMedia
        Consiste en verficar que los números generados tengan una media estadística igual a 0.5, de esta manera, se analiza la siguiente hipótesis:
        h0: mu = 0.5
        h1 : mu neq 0.5
        
        Esta prueba sigue los siguientes pasos:
        
        Paso 1: Calcular la media de los numeros generados
        paso 2: Calcular los limites superior e inferior
        Paso 3: Comparar si la media esta entre los limites superiores e inferiores
        
        Args:
            numeros (list): Arreglo que contiene los números generados de forma aleatoria
            valor_z (float): Valor z alpha/2 

        Returns:
            bool: Devuelve True or false dependiendo si pasa la prueba o no
        """
        n = len(numeros)
        numeros = np.array(numeros)
        mean = np.mean(numeros)
        # li = (1/2 - valor_z) * (1 / (12 * np.sqrt(n)))
        # ls = (1/2 + valor_z) * (1 / (12 * np.sqrt(n)))
        li = 0.5 - (valor_z / (np.sqrt(12 * n)))
        ls = 0.5 + (valor_z / (np.sqrt(12 * n)))

        # print('Resultado de 1/2 + valor z', 1/2 -valor_z, 'Resultado de 12sqrt(n)', (12 * np.sqrt(n)), 'resultado de 1/ el resultado anterior:',1/(12 * np.sqrt(n)))
        
        # print(f'valor del limite inferior: {li}\nvalor de la media: {mean}\nvalor del limite superior {ls}\nresultado boleado de li < mean < ls : {li < mean < ls}\nValor de N {n}')
        
        if li < mean< ls:
            return True
        else:
            return False    
    
```

```python
class MT19937:

	w, n = 32, 624
	f = 1812433253
	m, r = 397, 31
	a = 0x9908B0DF
	d, b, c = 0xFFFFFFFF, 0x9D2C5680, 0xEFC60000
	u, s, t, l = 11, 7, 15, 18

	def __init__(self, seed):
		self.X = [0] * MT19937.n
		self.cnt = 0
		self.initialize(seed)

	def initialize(self, seed):
		self.X[0] = seed
		for i in range(1, MT19937.n):
			self.X[i] = (MT19937.f * (self.X[i - 1] ^ (self.X[i - 1] >> (MT19937.w - 2))) + i) & ((1 << MT19937.w) - 1)
		self.twist()

	def twist(self):
		for i in range(MT19937.n):
			lower_mask = (1 << MT19937.r) - 1
			upper_mask =  (~lower_mask) & ((1 << MT19937.w) - 1)
			tmp = (self.X[i] & upper_mask) + (self.X[(i + 1) % MT19937.n] & lower_mask)
			tmpA = tmp >> 1
			if (tmp % 2):
				tmpA = tmpA ^ MT19937.a
			self.X[i] = self.X[(i + MT19937.m) % MT19937.n] ^ tmpA
		self.cnt = 0

	def temper(self):
		if self.cnt == MT19937.n:
			self.twist()
		y = self.X[self.cnt]
		y = y ^ ((y >> MT19937.u) & MT19937.d)
		y = y ^ ((y << MT19937.s) & MT19937.b)
		y = y ^ ((y << MT19937.t) & MT19937.c)
		y = y ^ (y >> MT19937.l)
		self.cnt += 1
		return y & ((1 << MT19937.w) - 1)

def main():
	rng = MT19937(0)
	for i in range(10):
		print(rng.temper())

# if __name__ == '__main__':
# 	main()
main()
```

### Pruebas de bondad de ajuste

También llamadas pruebas de aleatoridad. Las pruebas para números pseudoaletorios son pruebas estdísticas para decidir si una determinadda muestra o conjunto de datos responde a un patrón o se comportan de forma aleatoria.

Es importante asegurar que el generador usado busque una secuencia suficiente aleatoria, por ello es importante someter a estos generadores a dichas pruebas estadísticas.

Sim embargo este tipo de pruebas no son decisivas, debido a que un generador puede pasar una prueba estadística, pero si por ejemplo se cambia la semilla u otro segmento del ciclo, puede no pasar la misma prueba.
